```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import preprocessing

datos = pd.read_csv(r"datos_personas.csv")
datos.head(1)
```

## Representacion de los datos

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
ax1.set_title("Datos originales juntos")
ax1.plot(datos)

ax2.set_title("Ingreso")
ax2.plot(datos['ingreso'], linewidth = 0, marker="o", color = "red", markersize = 6)

ax3.set_title("Autos")
ax3.plot(datos['carros'], linewidth = 0, marker="+", color="orange", markersize = 16)
plt.show()
```

Como se puede apreciar en la grafica numero 1, el numero del salario de las personas se ve un cambio, pero en cuanto a los autos notamos que no existen cambios esto se debe que no estan escalados 

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.set_title("Ingreso")
ax1.plot(datos['ingreso'], linewidth = 0, marker="o", color = "red", markersize = 6)

ax2.set_title("Autos")
ax2.plot(datos['carros'], linewidth = 0, marker="+", color="orange", markersize = 16)

ax3.set_title("Distribucion de ingreso")
ax3.hist(datos["ingreso"], color='red',bins=200)

ax4.set_title("Distribucion de autos")
ax4.hist(datos['carros'], color='orange')


plt.show()
```

## Escalamiento con tensor flow 

El minmaxscaler toma en cuenta el valor mas alto y el valor mas bajo para realizar el modelado

```python
minmax = preprocessing.MinMaxScaler().fit_transform(datos)
minmax
```

Ahora tambien esta el normalizer

```python
normalizados = preprocessing.Normalizer().fit_transform(datos.T)
normalizados 
```
