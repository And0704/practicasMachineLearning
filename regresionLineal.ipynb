```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ingreso.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

Creacion de un scatter plot (grafico de dispersion para ver la data)

```python
plt.ylabel("Ingreso de ($)")
plt.xlabel("Promedio de hoars semanales trabajadas")
plt.scatter(x=df['horas'],y=df["ingreso"],color="red")
plt.show()
```

# Creacion del modelo de regresion lineal 

```python
from sklearn import  linear_model #Importa libreria para la regresion lineal
regresion = linear_model.LinearRegression() 
horas = df['horas'].values.reshape((-1,1))  #Para transformar en un vector que puede tomar el modelo
```

```python
modelo = regresion.fit(horas, df['ingreso']) #Ajustar la linea a los datos que se van a  enviar 
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[39], [40], [43],[44]]
modelo.predict(entrada)
```

```python
plt.xlabel("Horas trabajadas")
plt.ylabel("Ganacia de dinero")
plt.scatter(df["horas"],df["ingreso"],color='red')

plt.scatter(entrada,modelo.predict(entrada), color = 'orange')
plt.plot(entrada,modelo.predict(entrada), color = 'orange')
plt.show()
```
